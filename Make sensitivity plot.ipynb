```python
data_dir="./spectrometer_sensitivity_data/"

lbl={"firas":"FIRAS",
     "pristine":"PRISTINE",
     "pixie":"PIXIE",
     "super_pixie":"Super PIXIE",
     "voyage_2050":"Voyage 2050"
    }

figure()
for key in ["firas","pristine","pixie","super_pixie","voyage_2050"]:
    filename=data_dir + key + ".txt"
    x=np.loadtxt(filename)
    plot(x[:,0],x[:,1],label=lbl[key])
    legend()
    

legend(loc=0)
ylabel(r"$\Delta I_{\nu}$ Jy/sr")
xlabel("Frequency GHz")
xlim(5,9000) ; ylim(1e-2,1e7)
grid(which="both",linestyle="dashed",alpha=0.3)

loglog()
savefig("sensitivity.pdf",bbox_inches="tight")
```
